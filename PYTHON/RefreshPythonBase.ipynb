```python
##############################################
# SEARCHING ALGORITHMS
##############################################
# 1. Linear Search 
##############################################
# random order of elements
# 1. start from index 0 to len(arr)-1
# 2. compare each element with target
# 3. if found return index
# time complexity O(n)
# space complexity O(1)
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

##############################################
# 2. Binary Search
##############################################
# applied only on sorted array
# 1. start = 0, end = len(arr) -1
# 2. mid = (start + end) // 2   
# 3. compare arr[mid] with target
#   if arr[mid] == target: return mid
#   if arr[mid] < target: start = mid + 1
#   if arr[mid] > target: end = mid - 1
# while start <= end:repeat step 2 and 3
# if not found return -1
# time complexity O(log n)
# space complexity O(1)
def binary_search(arr, target):
    start = 0
    end = len(arr) - 1
    
    while start <= end:
        mid = (start + end) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            start = mid + 1
        else:
            end = mid - 1
    return -1
##############################################
```

```python
###############################################
# SORTING ALGORITHMS
###############################################
# 1. Bubble Sort
###############################################
# take adjacent elements and compare
# swap if not in order
# largest element bubbles to the right end in each pass
# After n-1 passes, we will be able to sort n elements
# time complexity O(n^2)
# space complexity O(1)
def bubble_sort(arr):
    n = len(arr) 
    for i in range(n): # i is passes
        for j in range(0, n-i-1): # last i elements are already sorted. so dont need to check. Optimization applied. Else repeatedly check is waste.
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

unsorted_list = []#[64, 34, 25, 12, 22, 11, 90]
sorted_list = bubble_sort(unsorted_list)
print("Bubble Sorted Elements are:", sorted_list)
################################################
# 2. Selection Sort
###############################################
# find the minimum element in unsorted array
# swap it with the first element (left) of unsorted array  
# repeat for remaining unsorted array n-1 times
# time complexity O(n^2)
# space complexity O(1)
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr   
               
unsorted_list = [64, 25, 12, 22, 11]
sorted_list = selection_sort(unsorted_list)
print("Selection Sorted Elements are:", sorted_list)
##############################################
# 3. Insertion Sort
##############################################
# divide the array into sorted and unsorted parts
# take one element from unsorted part       
# compare it with elements in sorted part
# insert it at the correct position in sorted part by comparing if its smaller than prevuous elements
# repeat until unsorted part is empty
# example Cards
# time complexity O(n^2)
# space complexity O(1)
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

unsorted_list = [12, 11, 13, 5, 6]
sorted_list = insertion_sort(unsorted_list)
print("Insertion Sorted Elements are:", sorted_list)
##############################################
# 4. Merge Sort
##############################################
# divide the array into 2 halves
# recursively sort each half
# merge the sorted halves
# time complexity O(n log n)
# space complexity O(n)
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

unsorted_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(unsorted_list) 
print("Merge Sorted Elements are:", sorted_list)
##############################################
# 5. Quick Sort
##############################################
# choose a pivot element
# partition the array around the pivot
# recursively sort the partitions
# time complexity O(n log n) average case
# space complexity O(log n)
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)
    
unsorted_list = [3,6,8,10,1,2,1]
sorted_list = quick_sort(unsorted_list)
print("Quick Sorted Elements are:", sorted_list)
print('#######################################################')
```

```python
# Inheritance
# like child get money from parents and not vice versa
print("################ SINGLE INHERITANCE ################\n")

# Parent class
class Parent:
    def __init__(self, parent_name):
        print("[Parent.__init__] Parent is getting ready üë®‚Äçü¶≥")
        self.parent_name = parent_name
        self.family_money = 10000   # attribute

    def give_advice(self):
        print("[Parent.give_advice] Parent giving advice...")
        return f"{self.parent_name} says: 'Always save money!'"

    def show_money(self):
        return f"Family Money ‚Üí ‚Çπ{self.family_money}"


# Child class
class Child(Parent):   # SINGLE INHERITANCE
    def __init__(self, child_name, parent_name):
        print("[Child.__init__] Child is loading... üßí")
        super().__init__(parent_name)   # call Parent constructor
        self.child_name = child_name
        self.pocket_money = 500         # attribute

    # overriding parent method
    def give_advice(self):
        print("[Child.give_advice] Child overriding advice...")
        return f"{self.child_name} says: 'I spend money faster than I get it! üòÇ'"

    def show_all_money(self):
        return (f"{self.show_money()}\n"
                f"{self.child_name}'s Pocket Money ‚Üí ‚Çπ{self.pocket_money}")


# Create object
kid = Child("Tommy", "Robert")

print("\n--- ADVICE TIME ---")
print(kid.give_advice())

print("\n--- MONEY STATUS ---")
print(kid.show_all_money())
```

```python
# MULTIPLE INHERITANCE
# like child get money from both parents
# Child class can inherit from more than one parent class
print('#######################################################')    

# Parent 1
class Mom:
    def __init__(self, mom_name):
        print("[Mom.__init__] Mom is getting ready üë©‚Äçü¶∞")
        self.mom_name = mom_name        # attribute
        self.mom_skill = "Cooking Pasta üçù"

    def show_mom_skill(self):
        return f"{self.mom_name}'s Skill ‚Üí {self.mom_skill}"


# Parent 2
class Dad:
    def __init__(self, dad_name):
        print("[Dad.__init__] Dad is getting ready üë®")
        self.dad_name = dad_name        # attribute
        self.dad_skill = "Expert Driving üöóüí®"

    def show_dad_skill(self):
        return f"{self.dad_name}'s Skill ‚Üí {self.dad_skill}"


# Child inherits from BOTH Mom and Dad
class Kid(Mom, Dad):
    def __init__(self, kid_name, mom_name, dad_name):
        print("[Kid.__init__] Kid is loading... üßí")

        # call parents' constructors manually
        Mom.__init__(self, mom_name)
        Dad.__init__(self, dad_name)

        self.kid_name = kid_name        # attribute
        self.kid_skill = "Eating pasta while Dad drives üòÇ"

        print("[Kid.__init__] Kid is fully ready! üéâ")

    def show_all_skills(self):
        return (
            f"Kid: {self.kid_name}\n"
            f" - {self.show_mom_skill()}\n"
            f" - {self.show_dad_skill()}\n"
            f" - Kid's Skill ‚Üí {self.kid_skill}"
        )


# Create an instance (object)
little_kid = Kid("Lovely", "Saly", "Davy")

print("\n--- FAMILY SKILLS ---")
print(little_kid.show_all_skills())
```

```python
print("\n################## POLYMORPHISM WITH ABSTRACT BASE CLASSES ##################\n")

from abc import ABC, abstractmethod

# ============================
# ABSTRACT BASE CLASS (INTERFACE)
# ============================
class FamilyMember(ABC):
    
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def wake_up(self):
        pass   # Must be implemented by children

    @abstractmethod
    def daily_task(self):
        pass   # Must be implemented by children

    # Common method (not abstract)
    def introduce(self):
        return f"Hi! I am {self.name}."


# ============================
# CHILD 1 ‚Äî DAD
# ============================
class Dad(FamilyMember):

    def __init__(self, name, favorite_coffee):
        super().__init__(name)
        self.favorite_coffee = favorite_coffee

    def wake_up(self):
        return f"{self.name}: Wakes up at 6 AM and brews {self.favorite_coffee} ‚òï"

    def daily_task(self):
        return f"{self.name}: Fixes something in the house üîß"


# ============================
# CHILD 2 ‚Äî MOM
# ============================
class Mom(FamilyMember):

    def __init__(self, name, morning_routine):
        super().__init__(name)
        self.morning_routine = morning_routine

    def wake_up(self):
        return f"{self.name}: Wakes up and starts {self.morning_routine} üßπ"

    def daily_task(self):
        return f"{self.name}: Prepares breakfast and organizes the day üç≥"


# ============================
# CHILD 3 ‚Äî KID
# ============================
class Kid(FamilyMember):

    def __init__(self, name, favorite_cartoon):
        super().__init__(name)
        self.favorite_cartoon = favorite_cartoon

    def wake_up(self):
        return f"{self.name}: Wakes up only if someone turns off the Wi-Fi üò¥üìµ"

    def daily_task(self):
        return f"{self.name}: Watches {self.favorite_cartoon} and makes a mess everywhere üòÇ"


# ============================
# POLYMORPHISM IN ACTION
# ============================
family = [
    Dad("John", "Espresso"),
    Mom("Maria", "Morning Yoga"),
    Kid("Leo", "Peppa Pig")
]

for person in family:
    print(person.introduce())
    print(person.wake_up())
    print(person.daily_task())
    print()
```

```python
print("################## ENCAPSULATION ##################")
# Encapsulation = keeping some data private so nobody can misuse it.
# Just like every family has public info and private secrets, a class can have:
# Public attributes ‚Üí Anyone can access
# Private attributes ‚Üí Hidden, locked inside the class
# __private_secret has two underscores ‚Üí meaning no outsider can access it directly.You can only access it through a method.
# ‚û°Ô∏è Encapsulation = controlling who can see what.

class FamilySecret:
    def __init__(self):
        self.public_info = "We are a happy family üôÇ"
        self.__private_secret = "Mom hides chocolates on the top shelf üç´ü§´"

    def reveal_secret(self):
        return f"Revealed Secret ‚Üí {self.__private_secret}"

family = FamilySecret()

print(family.public_info)
print(family.reveal_secret())
# print(family.__private_secret)  # ‚ùå ERROR: encapsulated / private

print("\n################## INHERITANCE ##################")
# Inheritance means a child automatically gets features from the parent.
# In real life  A kid inherits:
# Looks üëÄ
# Skills üé®
# Behaviors üòÇ  from parents.
# Kid inherits mom‚Äôs skills
# Kid automatically gets mom_skill and show_skill()
# Kid adds more skills of its own
# ‚û°Ô∏è Inheritance = reusing parent‚Äôs features without rewriting code.
class Mom:
    def __init__(self):
        self.mom_skill = "Making the BEST pasta üçù"

    def show_skill(self):
        return f"Mom's Skill ‚Üí {self.mom_skill}"


class Kid(Mom):  # inherits from Mom
    def __init__(self):
        super().__init__()
        self.kid_skill = "Eating pasta like a legend üòÇ"

    def show_kid_skill(self):
        return f"Kid's Skill ‚Üí {self.kid_skill}"

k = Kid()
print(k.show_skill())
print(k.show_kid_skill())

print("\n################## POLYMORPHISM ##################")
# Polymorphism = same function name, different behaviors.
# In real life Everyone says "Good morning", but differently:
# Mom: ‚ÄúGood morning sweetie üòò‚Äù
# Dad: ‚ÄúMorning champ üí™‚Äù
# Kid: ‚ÄúUghh... 5 more minutes üò¥‚Äù
# Each family member overrides greet()
# Same function ‚Üí different output.
# ‚û°Ô∏è Polymorphism = same action, different behavior.
class Person:
    def greet(self):
        return "Hello!"

class Mom(Person):
    def greet(self):
        return "Mom says: Good morning sweetie üòò"

class Dad(Person):
    def greet(self):
        return "Dad says: Morning champ! üí™"

class Kid(Person):
    def greet(self):
        return "Kid says: Ughhh‚Ä¶ 5 more minutes üò¥"

def greet_everyone(person):
    print(person.greet())

greet_everyone(Mom())
greet_everyone(Dad())
greet_everyone(Kid())

print("\n################## POLYMORPHISM using METHOD OVERLOADING ##################\n")
# Same method name ‚Üí works differently depending on arguments
# Python ‚Äúfakes‚Äù overloading using default parameters
class Mom:
    def greet(self, name=None):
        if name:
            return f"Mom: Good morning, {name}! üòò"
        return "Mom: Good morning sweetie üòò"


class Dad:
    def greet(self, mood=None):
        if mood == "angry":
            return "Dad: Morning‚Ä¶ who finished the snacks? üò°"
        elif mood == "happy":
            return "Dad: Good morning champ! üòÑ"
        return "Dad: Morning! üí™"


class Kid:
    def greet(self, time=None):
        if time == "early":
            return "Kid: Nooooo‚Ä¶ too early üò©"
        elif time == "late":
            return "Kid: OMG I'm gonna be late!! üò±"
        return "Kid: Ugh‚Ä¶ morning already? üò¥"


# Same method name ‚Üí different arguments ‚Üí different behaviors
mom = Mom()
dad = Dad()
kid = Kid()

print(mom.greet())
print(mom.greet("Lovely"))

print(dad.greet())
print(dad.greet("happy"))
print(dad.greet("angry"))

print(kid.greet())
print(kid.greet("early"))
print(kid.greet("late"))


from abc import ABC, abstractmethod

print("\n################## POLYMORPHISM using ABSTRACT BASE CLASS ##################\n")
# Same function name ‚Üí different behaviors
# Ensures EVERY child class must implement the method
# Person is abstract ‚Üí cannot be instantiated
# All child classes must override greet()
# Each class gives its own version of the same method
# abstract method = a method that is declared, but contains no implementation.it is a decorator that is used to declare a method as abstract within an abstract base class.
# flexibility and integration in code design.
# a single function can work in different ways depending on the object that calls it.

# Abstract Base Class
class Person(ABC):
    
    @abstractmethod
    def greet(self):
        pass   # Force each subclass to implement greet()


# Child 1
class Mom(Person):
    def __init__(self):
        self.role = "Mom"
    
    def greet(self):
        return "Mom says: Good morning sweetie üòò"

    def cook(self):
        return "Mom is cooking pancakes ü•û"


# Child 2
class Dad(Person):
    def __init__(self):
        self.role = "Dad"
    
    def greet(self):
        return "Dad says: Morning champ! üí™"

    def drive(self):
        return "Dad is warming up the car üöóüí®"


# Child 3
class Kid(Person):
    def __init__(self):
        self.role = "Kid"
    
    def greet(self):
        return "Kid says: Ughhh‚Ä¶ 5 more minutes üò¥"

    def play(self):
        return "Kid is playing video games üéÆ"


# create object .
# Polymorphism in action
family = [Mom(), Dad(), Kid()]

for person in family:
    print(person.greet())    # same function ‚Üí different behaviors


print("\n################## ABSTRACTION ##################")
# Abstraction = show only what‚Äôs necessary, hide complexity.
# In real life  You don‚Äôt need to know:
# How Dad pays electricity bill
# How Mom pays school fees
# You just see the final action: "Bill Paid".
# Every family member must implement their own version of pay().
# ‚û°Ô∏è Abstraction = forcing children to follow a rule without explaining the details.
from abc import ABC, abstractmethod

class BillPayment(ABC):
    @abstractmethod
    def pay(self):
        pass

class DadPayment(BillPayment):
    def pay(self):
        return "Dad paid the electricity bill üí°"

class MomPayment(BillPayment):
    def pay(self):
        return "Mom paid the school fees üéí"

def process_payment(payment):
    print(payment.pay())

process_payment(DadPayment())
process_payment(MomPayment())

# | Concept           | Real-life Meaning            | Code Meaning                         |
# | ----------------- | ---------------------------- | ------------------------------------ |
# | **Encapsulation** | Family secrets locked away   | Private variables (`__secret`)       |
# | **Inheritance**   | Kids get parents‚Äô qualities  | Child class gets parent methods      |
# | **Polymorphism**  | Everyone greets differently  | Same function name, different output |
# | **Abstraction**   | You see results, not process | Abstract methods (must override)     |

```

```python
print("\n################## ENCAPSULATION ‚Äî SEEING PUBLIC / PROTECTED / PRIVATE ##################\n")
# Private variable cannot be accessed directly from outside the class even for the child classes. Using getter and setter methods we can access and modify private variables inside the class.
# Protected variable can be accessed by child classes but not outside.
# Public variable can be accessed from anywhere.
class Mom:
    def __init__(self):
        self.public_var = "I am public üòä"
        self._protected_var = "I am protected ü§´"
        self.__private_var = "I am private üîê"

mom = Mom()

print("---- dir(mom) OUTPUT (to see how attributes appear) ----")
print(dir(mom))

print("\nExplanation:")
print("‚Ä¢ public_var ‚Üí appears normally")
print("‚Ä¢ _protected_var ‚Üí appears with one underscore (convention)")
print("‚Ä¢ __private_var ‚Üí becomes _Mom__private_var because of name-mangling")

# =====================================================================
#            FULL FAMILY ENCAPSULATION EXAMPLE WITH GETTERS
# =====================================================================

class Mom:
    def __init__(self, family_name):
        self.family_name = family_name          # PUBLIC
        self._recipe = "Special Kizhanghuuzhukku with meencurry üçõ"        # PROTECTED
        self.__atm_pin = "1234"                 # PRIVATE

    # ----------------------- GETTERS -----------------------

    def get_family_name(self):
        return f"Family Name (Public): {self.family_name}"

    def get_protected_recipe(self):
        return f"Protected Recipe: {self._recipe}"

    def get_private_pin(self):# i just expose get_private_pin method to outside without exposing private variable directly
        return f"Private PIN: {self.__atm_pin}"

    # ----------------------- SETTER ------------------------

    def set_atm_pin(self, new_pin):
        if len(new_pin) == 4 and new_pin.isdigit():
            self.__atm_pin = new_pin
            return "Mom: PIN updated ‚úî"
        return "Mom: Invalid PIN format ‚ùå"

    def show_info(self):
        return (f"Public attribute ‚Üí {self.family_name}\n"
                f"Protected attribute ‚Üí {self._recipe}\n"
                f"Private attribute ‚Üí Cannot access directly")


class Kid(Mom):
    def reveal_recipe(self):
        return f"Kid: I know mom's protected recipe ‚Üí {self._recipe}"

    def try_access_pin(self):
        try:
            return self.__atm_pin
        except:
            return "Kid: I cannot access mom‚Äôs private PIN! üò≠"


# ----------- DEMO -----------
print("\n################## ENCAPSULATION FAMILY DEMO ##################\n")

mom = Mom("Davy Family")
kid = Kid("Davy Family")

print("---- Mom's Info ----")
print(mom.show_info())

print("\n---- Using NEW GETTERS ----")
print(mom.get_family_name())
print(mom.get_protected_recipe())
print(mom.get_private_pin())

print("\n---- Kid accessing protected variable ----")
print(kid.reveal_recipe())

print("\n---- Kid trying to access private variable ----")
print(kid.try_access_pin())

print("\n---- Using getter for private PIN ----")
print(mom.get_private_pin())

print("\n---- Updating private PIN ----")
print(mom.set_atm_pin("9876"))
print(mom.get_private_pin())

print("\n################## END ##################")
```

```python
print("\n################## ABSTRACTION (1 BASE + 1 CHILD) ##################\n")
# abstrcation is the concept of hiding the complex implementation details and showing only the essential features of the object.
# It helps in reducing programming complexity and effort.
# In real life You don‚Äôt need to know:
# How washing machine works internally
# You just press "Start" and it works.
# Every appliance must implement turn_on().
# ‚û°Ô∏è Abstraction = focusing on what an object does instead of how it does it.

from abc import ABC, abstractmethod

# =====================================================================
#                      ABSTRACT BASE CLASS
# =====================================================================

class Appliance(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def turn_on(self):
        """Abstract method - must be implemented by child class"""
        pass

    def info(self):
        """Normal method - works without overriding"""
        return f"This is an appliance made by {self.brand}."


# =====================================================================
#                      CHILD CLASS (CONCRETE)
# =====================================================================

class WashingMachine(Appliance):
    def __init__(self, brand, load_capacity):
        super().__init__(brand)
        self.load_capacity = load_capacity

    def turn_on(self):
        return f"{self.brand} washing machine is now ON. Washing {self.load_capacity}kg of clothes üß∫‚ö°"


# =====================================================================
#            FUNCTION THAT USES BASE-CLASS REFERENCE 
# =====================================================================

def run_appliance(appliance: Appliance):
    """Takes ANY Appliance (base class reference)"""
    print(appliance.info())
    print(appliance.turn_on())


# =====================================================================
#                          DEMO
# =====================================================================

wm = WashingMachine("Samsung", 7)  # object of child class

run_appliance(wm)  # Passed as Appliance type

print("\n################## END ##################")
```

```python
print("\n################## ABSTRACTION ‚Äî FAMILY EXAMPLE ##################\n")

from abc import ABC, abstractmethod

# =====================================================================
#                     ABSTRACT BASE CLASS (ABC)
# =====================================================================

class FamilyMember(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def morning_routine(self):
        """Every family member has a morning routine, but done differently."""
        pass

    @abstractmethod
    def responsibility(self):
        """Each member has a responsibility, but details differ."""
        pass

    # Concrete method (shared logic)
    def introduce(self):
        return f"Hi, I'm {self.name} from the family!"


# =====================================================================
#                        DERIVED CLASSES
# =====================================================================

class Mom(FamilyMember):
    def __init__(self, name):
        super().__init__(name)
        self.coffee_ready = True

    def morning_routine(self):
        return "Mom: Wakes up early, makes breakfast and coffee ‚òïüç≥"

    def responsibility(self):
        return "Mom: Organizes the day and keeps everyone on track üìÖ"


class Dad(FamilyMember):
    def __init__(self, name):
        super().__init__(name)
        self.news_read = False

    def morning_routine(self):
        return "Dad: Reads the news and complains about bills üóûÔ∏èüí∏"

    def responsibility(self):
        return "Dad: Takes care of car, repairs, and weekend BBQ üîßüî•"


class Kid(FamilyMember):
    def __init__(self, name):
        super().__init__(name)
        self.homework_done = False

    def morning_routine(self):
        return "Kid: Snoozes alarm, asks for 5 more minutes üò¥‚è∞"

    def responsibility(self):
        return "Kid: Go to school, do homework (sometimes) üìöüòÖ"


# =====================================================================
#                        DEMO (POLYMORPHISM + ABSTRACTION)
# =====================================================================

family = [
    Mom("Saly"),
    Dad("Davy"),
    Kid("Lovely")
]

for member in family:
    print(member.introduce())
    print(member.morning_routine())
    print(member.responsibility())
    print("------------------------------------------------")

print("\n################## END ##################")
```

```python
# Magic Methods
# special methods with double underscores that allow customization of class behavior for built-in operations.
# dunder methods : methods that have double underscores at the beginning and end of their names.
# __init__ : Initializes a new instance of a class (constructor).
# __str__ : Returns a readable string representation of an object.
# __repr__ : Returns an official string representation of an object (for debugging).
# __add__ : Defines behavior for the + operator.
# __len__ : Returns the length of an object.
# __setitem__ : Defines behavior for assigning a value to an index/key.Sets an item in a container.
# __getitem__ : Defines behavior for accessing a value by index/key. Gets an item from a container.
# __delitem__ : Defines behavior for deleting an item by index/key. Deletes an item from a container.
# __contains__ : Defines behavior for the in operator.Checks if an item is in a container.
# __call__ : Allows an instance to be called as a function.
# __eq__ : Defines behavior for the == operator.

print("################## MAGIC METHODS ##################\n")  

# Example 1: Default magic methods from object
class Person:
    pass

Lovely = Person()
print("Default magic methods for Person object:")
print(dir(Lovely))  # shows inherited dunder methods
print("\n")

# Example 2: Custom magic methods
class FamilyMember:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.attributes = {}  # For __setitem__ and __getitem__

    def __str__(self):  # user-friendly string
        return f"FamilyMember(Name: {self.name}, Age: {self.age})"

    def __repr__(self):  # developer/debug-friendly string
        return f"FamilyMember(name='{self.name}', age={self.age})"

    def __add__(self, other):  # behavior for '+'
        return f"{self.name} and {other.name} are now friends! ü§ù"

    def __len__(self):  # behavior for len()
        return self.age

    def __setitem__(self, key, value):  # set item like in a dict
        self.attributes[key] = value

    def __getitem__(self, key):  # get item like in a dict
        return self.attributes[key]

# Create objects
mom = FamilyMember("Saly", 40)
dad = FamilyMember("Davy", 45)

# __str__ and __repr__
print(mom)            # __str__
print(repr(mom))      # __repr__

# __add__
print(mom + dad)      # __add__

# __len__
print(f"Length of mom's age: {len(mom)} years")  # __len__

# __setitem__ and __getitem__
mom["hobby"] = "Gardening"  # __setitem__
mom["city"] = "Paris"       # __setitem__
print(f"Mom's hobby: {mom['hobby']}")  # __getitem__
print(f"Mom lives in: {mom['city']}")  # __getitem__

print("\n################## END ##################")
```

```python
# Operator Overloading
# customizing the behavior of standard operators (+, -, *, etc.) for user-defined classes.
# __add__ : Defines behavior for the + operator.
# __sub__ : Defines behavior for the - operator.
# __mul__ : Defines behavior for the * operator.
# __truediv__ : Defines behavior for the / operator.
# __eq__ : Defines behavior for the == operator.
# __lt__ : Defines behavior for the < operator.
# __gt__ : Defines behavior for the > operator.
# __le__ : Defines behavior for the <= operator.
# __ge__ : Defines behavior for the >= operator.
# __ne__ : Defines behavior for the != operator.

print("################## MAGIC METHODS: ARITHMETIC & COMPARISON ##################\n")

class FamilyMember:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.attributes = {}  # for __setitem__ and __getitem__

    # String representations
    def __str__(self):
        return f"{self.name} ({self.age} yrs)"

    def __repr__(self):
        return f"FamilyMember(name='{self.name}', age={self.age})"

    # Arithmetic operators
    def __add__(self, other):
        return f"{self.name} and {other.name} are now friends! ü§ù"

    def __sub__(self, other):
        return abs(self.age - other.age)

    def __mul__(self, other):
        return self.age * other.age

    # Comparison operators
    def __eq__(self, other):
        return self.age == other.age

    def __lt__(self, other):
        return self.age < other.age

    def __le__(self, other):
        return self.age <= other.age

    def __gt__(self, other):
        return self.age > other.age

    def __ge__(self, other):
        return self.age >= other.age

    # Length
    def __len__(self):
        return self.age

    # Item access
    def __setitem__(self, key, value):
        self.attributes[key] = value

    def __getitem__(self, key):
        return self.attributes[key]

# Create objects
mom = FamilyMember("Saly", 55)
dad = FamilyMember("Davy", 63)
aunt = FamilyMember("Lovely", 29)

# String representations
print(mom)        # __str__
print(repr(dad))  # __repr__

# Arithmetic operations
print(mom + dad)   # __add__
print(dad - mom)   # __sub__ (age difference)
print(mom * aunt)  # __mul__ (age multiplication)

# Comparison operations
print(mom == aunt)  # True (__eq__)
print(mom < dad)    # True (__lt__)
print(dad > mom)    # True (__gt__)
print(mom <= aunt)  # True (__le__)
print(dad >= aunt)   # True (__ge__)

# Length
print(f"Mom's age as length: {len(mom)}")

# Item access
mom["hobby"] = "Gardening"
mom["city"] = "Paris"
print(f"Mom's hobby: {mom['hobby']}")
print(f"Mom lives in: {mom['city']}")
print("\n################## END ##################")

print("################## OPERATOR OVERLOADING ##################\n")
class PocketMoney:
    def __init__(self, amount):
        self.amount = amount

    def __add__(self, other):
        return PocketMoney(self.amount + other.amount)

    def __str__(self):
        return f"Total Pocket Money: ‚Çπ{self.amount}"
mom_money = PocketMoney(50)
dad_money = PocketMoney(100)
print(mom_money + dad_money)   # Operator polymorphism
print("\n################## END ##################")

```

```python
print("################## COMPLEX NUMBER OPERATOR OVERLOADING ##################\n")

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    # String representation
    def __str__(self):
        return f"{self.real} + {self.imag}i"

    def __repr__(self):
        return f"ComplexNumber({self.real}, {self.imag})"

    # Addition
    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    # Subtraction
    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)

    # Multiplication
    def __mul__(self, other):
        r = self.real * other.real - self.imag * other.imag
        i = self.real * other.imag + self.imag * other.real
        return ComplexNumber(r, i)

    # Division
    def __truediv__(self, other):
        denom = other.real**2 + other.imag**2
        r = (self.real * other.real + self.imag * other.imag) / denom
        i = (self.imag * other.real - self.real * other.imag) / denom
        return ComplexNumber(r, i)

# Create complex numbers
c1 = ComplexNumber(3, 2)
c2 = ComplexNumber(1, 7)

# Arithmetic operations
print(f"c1: {c1}")
print(f"c2: {c2}")
print(f"c1 + c2 = {c1 + c2}")  # addition
print(f"c1 - c2 = {c1 - c2}")  # subtraction
print(f"c1 * c2 = {c1 * c2}")  # multiplication
print(f"c1 / c2 = {c1 / c2}")  # division

print("\n################## END ##################")
```

```python
print("\n################## CUSTOM EXCEPTIONS ##################\n")

# ------------------------------------------------------------
# WHY CUSTOM EXCEPTIONS?
# ------------------------------------------------------------
# Python has built-in exceptions (ValueError, TypeError, ZeroDivisionError‚Ä¶)
# but sometimes YOU want to represent your own error types.
#
# Example: In a family-management app, you may want errors like:
# - MissingChoreError
# - LateDinnerError
#
# Custom exceptions help create MORE MEANINGFUL and SPECIFIC error messages.
# ------------------------------------------------------------


# ------------------------------------------------------------
# BASE EXCEPTION CLASS
# ------------------------------------------------------------
# You create a base class so all family-related errors share a common ancestor.
# This allows grouping them in exception handling.
class FamilyError(Exception):
    """Base class for all family-related custom exceptions."""
    pass


# ------------------------------------------------------------
# CUSTOM EXCEPTION 1: MissingChoreError
# ------------------------------------------------------------
# This error is raised when a family member fails to complete a chore.
class MissingChoreError(FamilyError):
    """Raised when a family member misses their assigned chore."""
    
    def __init__(self, member_name, chore):
        self.member_name = member_name
        self.chore = chore
        
        # The error message passed to Exception
        super().__init__(f"{self.member_name} missed their chore: {self.chore}")


# ------------------------------------------------------------
# CUSTOM EXCEPTION 2: LateDinnerError
# ------------------------------------------------------------
# This error is raised when dinner is served later than expected.
class LateDinnerError(FamilyError):
    """Raised when dinner is served late."""
    
    def __init__(self, delay_minutes):
        self.delay_minutes = delay_minutes
        
        # Provides a meaningful message
        super().__init__(f"Dinner is late by {self.delay_minutes} minutes!")


# ------------------------------------------------------------
# FUNCTIONS THAT MAY RAISE CUSTOM EXCEPTIONS
# ------------------------------------------------------------

def perform_chore(member_name, chore_done):
    """Check if a chore is done; if not, raise MissingChoreError."""
    if not chore_done:
        raise MissingChoreError(member_name, "Taking out the trash")


def serve_dinner(delay_minutes):
    """Serve dinner; if delayed, raise LateDinnerError."""
    if delay_minutes > 0:
        raise LateDinnerError(delay_minutes)


# ------------------------------------------------------------
# USING AND HANDLING CUSTOM EXCEPTIONS
# ------------------------------------------------------------
# We wrap calls in try/except to catch and handle custom errors gracefully.


# 1. Testing chore performance
try:
    perform_chore("Jake", False)  # This will raise MissingChoreError
except MissingChoreError as e:
    print(f"Custom Exception Caught: {e}")


# 2. Testing dinner delay
try:
    serve_dinner(15)  # This will raise LateDinnerError
except LateDinnerError as e:
    print(f"Custom Exception Caught: {e}")


print("\n################## END ##################")
```

```python
# Implementation of a Custom List Class
import ctypes # to implement a python reference array

class CustomList:
    def __init__(self):
        initial_capacity = 1
        self.capacity = initial_capacity # how much we can store
        self.size = 0                    # how much are stored right now
        self.array = self.__create_array(self.capacity) # private array
    
    def __create_array(self, capacity):
        # create a new referential array with given capacity
        return (capacity * ctypes.py_object)()
    
    def __resize(self, new_capacity):
        new_array = self.__create_array(new_capacity)
        for i in range(self.size):
            new_array[i] = self.array[i]
            
        self.array = new_array #replace old array
        self.capacity = new_capacity        
    # 1. Implementing append method
    def append(self, item):
        if self.size == self.capacity:
            self.__resize(2 * self.capacity) # double the capacity
            
        self.array[self.size] = item
        self.size += 1
    # 2. Implementing __len__ and __str__ methods    
    def __len__(self):
        return self.size
    
    def __str__(self):
        output = ''
        for i in range(self.size):
            output += str(self.array[i]) + ','
        return '[' + output[:-1] +']'  # remove last comma
    
    # 3. implementing pop method : removes and returns the last item
    def pop(self):
        if self.size == 0:
            raise IndexError("pop from empty list")
        
        popped_item = self.array[self.size - 1] # get last item
        self.size -= 1        
        return  popped_item # return last item


myList = CustomList()
myList.append(10)
myList.append(20)
myList.append(30)
myList.append(40)
print(myList)
print("Custom List Size:", len(myList))  # Output: 4
myList.pop()
print(myList)
print("Custom List Size after pop:", len(myList))  # Output: 3
myList.pop()
print(myList)
print("Custom List Size after pop:", len(myList))  # Output: 2
myList.pop()
print(myList)
print("Custom List Size after pop:", len(myList))  # Output: 1
myList.pop()
print(myList)   
print("Custom List Size after pop:", len(myList))  # Output: 0  
myList.pop()  # This will raise IndexError

```
