```python
# 242 Valid Anagram
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s)!= len(t):
            return False
        countS, countT = {},{}
        # create hashmaps
        for i in range(len(s)):
            countS[s[i]] = 1 + countS.get(s[i],0)
            countT[t[i]] = 1+ countT.get(t[i],0)
        # compare hashmaps
        for c in countS: # iterate through keys. c is the key
            if countS[c] != countT.get(c,0): # if count at character c is same in s & t
                return False
        return True
        # return Counter(s) == Counter(t)
        # Time : O(s+t)
        # Space: O(n)

        # return sorted(s) == sorted(t)
        # sorting Time : O(nlogn), O(n**2)
        # sorting space : O(1),O(n)
        
```

```python
# 1. Two Sum
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i, num in enumerate(nums):  # i = index, num = value
            diff = target - num
            if diff in seen:
                return [seen[diff], i]  # return the index of the pair
            # saving the number as a key and its position (index) as the value inside the dictionary seen.
            seen[num] = i # Remember that I saw the number num at index i

"""
Approach    Code	        Time	Space	Notes
BruteForce	Double loop	    O(n²)	O(1)	Simple but slow
Hash Map	seen[num] = i	O(n)	O(n)	Fastest and most common
"""

```

```python
#  Example dictionary
seen = {2: 0, 7: 1, 11: 2, 15: 3}

# 1️. Get value from key
print("Value for key 7:", seen[7])   # direct lookup
print("Value for key 11 (using get):", seen.get(11))   # safe lookup
print("Value for missing key 20 (default=not found):", seen.get(20, "not found"))

# 2️. Get all keys
print("\nAll keys:", list(seen.keys()))

# 3️. Get all values
print("All values:", list(seen.values()))

# 4️. Get both key and value pairs
print("\nAll key-value pairs:")
for key, value in seen.items():
    print(f"Key = {key}, Value = {value}")

# 5️. Find key from a value (manual search)
search_value = 2
for key, value in seen.items():
    if value == search_value:
        print(f"\nKey for value {search_value} is: {key}")
        break
```
