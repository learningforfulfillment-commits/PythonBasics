SPEAKING
---------

SICH VORSTELLEN (INTRODUCING YOURSELF)
-----------------------------------------
Guten Tag, mein Name ist ____.
Ich bin ____ Jahre alt.
Ich komme aus ____ und wohne seit ____ Jahren in Deutschland.
Ich arbeite als ____ bei ____.
In meiner Freizeit lese ich gern, treibe Sport und treffe Freunde.
Mein Ziel ist es, mein Deutsch weiter zu verbessern.

MEIN ALLTAG (MY DAILY ROUTINE)
-------------------------------
Unter der Woche stehe ich meistens um ____ Uhr auf.
Danach frühstücke ich und fahre zur Arbeit.
Ich arbeite von ____ Uhr bis ____ Uhr.
Am Abend koche ich, schaue fern oder lerne Deutsch.
Am Wochenende mache ich oft einen Ausflug oder treffe Freunde.

EINLADEN UND PLANEN (INVITING AND PLANNING)
-------------------------------------------
Möchtest du am Samstag zu mir kommen?
Hast du am Wochenende Zeit?
Sollen wir uns um ____ Uhr treffen?
Wo wollen wir uns treffen?
Ich schlage vor, dass wir ins Kino gehen.
Es wäre besser, wenn wir uns im Café treffen.

MEINUNGEN SAGEN (GIVING OPINIONS)
----------------------------------
Ich denke, dass das eine gute Idee ist.
Meiner Meinung nach ist das sehr wichtig.
Ich finde, dass Sport gesund und wichtig ist.
Ich bin damit einverstanden.
Ich bin anderer Meinung.
Ich bin nicht sicher, ob das funktioniert.

VORSCHLÄGE MACHEN (MAKING SUGGESTIONS)
-----------------------------------------
Sollen wir vielleicht einen Spaziergang machen?
Wie wäre es, wenn wir zusammen kochen?
Wir könnten auch im Park ein Picknick machen.
Vielleicht könnten wir später noch etwas trinken gehen.
Ich schlage vor, dass wir heute früher anfangen.

PROBLEME UND BESCHWERDEN (PROBLEMS AND COMPLAINTS)
---------------------------------------------------
Ich habe ein Problem mit der Wohnung.
Es gibt ein Problem mit der Heizung.
Ich bin unzufrieden mit dem Service.
Könnten Sie das bitte reparieren?
Könnten Sie mir bitte helfen?
Ich brauche eine Lösung für dieses Problem.

NACHFRAGEN UND KLARHEIT (ASKING FOR CLARIFICATION)
--------------------------------------------------
Könnten Sie das bitte wiederholen?
Könnten Sie bitte etwas langsamer sprechen?
Ich habe das nicht ganz verstanden.
Was meinen Sie genau?
Könnten Sie das näher erklären?

ZUKUNFT UND PLÄNE (FUTURE AND PLANS)
----------------------------------------
Ich habe vor, nächstes Jahr mehr Deutsch zu lernen.
In Zukunft möchte ich eine bessere Arbeitsstelle finden.
Ich möchte später eine eigene Wohnung kaufen.
Vielleicht werde ich noch eine Ausbildung machen.
Ich hoffe, dass ich die Prüfung bestehe.

VERGLEICHEN UND ABWÄGEN (COMPARING OPTIONS)
----------------------------------------------------
Auf der einen Seite ist das billiger, aber auf der anderen Seite ist es nicht so bequem.
Mit dem Auto ist man schneller, aber der Zug ist umweltfreundlicher.
Ich denke, dass Variante A besser für uns ist.
Für mich ist wichtig, dass es nicht zu teuer ist.

GESPRÄCH BEENDEN (ENDING THE CONVERSATION)
-----------------------------------------------
Vielen Dank für das Gespräch.
Es hat mich gefreut, mit Ihnen zu sprechen.
Ich hoffe, wir sehen uns bald wieder.
Dann sind wir uns einig.
Auf Wiedersehen und einen schönen Tag noch.

1. Arbeit/Beruf
--------------

- speaking (Teil 1 intro, Teil 2 opinion, Teil 3 planning) 
- writing (new job, problems, internship).
- Points to cover: what job, daily tasks, pros/cons, colleagues/boss, future plans


Speaking
----------
Teil 1: Kennenlernen (introduce + 1 follow-up)
------------------------------------------------
Expected questions: "Was machen Sie beruflich?" / "Wo arbeiten Sie?" / Follow-up
------------------------------------------------------------------------------------
Hallo, mein Name ist Lovely, und ich komme aus Indien.
Ich wohne seit 2024 in Landshut in der Nähe von München, weil ich oft von zu Hause aus arbeite. Die Miete dort ist günstiger als in München, und die Zug- und Busverbindungen sind sehr gut.
Ich arbeite als erfahrener Data Science Berater in einer Firma in München. Jeden Tag analysiere ich Daten mit Python und erstelle Modelle.
Am Vormittag checke ich E-Mails, und am Nachmittag habe ich Meetings mit meinen Kollegen, meistens online.
Einerseits mag ich den Job, weil er kreativ ist, ich viel lerne und ich gut verdiene. Andererseits ist er stressig wegen der vielen Fristen und weil ich lange am Computer sitze.
In Zukunft möchte ich mich mehr auf Künstliche Intelligenz spezialisieren.
Was machen Sie beruflich? Wie gefällt Ihnen München?
München ist toll, die Stadt ist modern und hat gute Verbindungen zur Arbeit.

Teil 2: Thema besprechen (~2 min solo)
--------------------------------------------
Card topic: "Mein Beruf" – Cover: What/Tasks, daily routine, pros/cons, future, question to partner.
Was machen Sie genau als Data Science Berater?
Ich analysiere Daten mit Python, baue Machine-Learning-Modelle und erstelle Berichte. Zum Beispiel suche ich Muster in großen Datenmengen.
Arbeiten Sie im Büro oder von zu Hause?
Meist arbeite ich von zu Hause in Landshut, nur zweimal im Jahr fahre ich nach München ins Büro.
Wie lange arbeiten Sie schon in diesem Beruf?
Seit drei Jahren; ich habe viel dazugelernt, besonders Machine Learning.
Gefällt Ihnen der Job? Warum?
Ja, er ist kreativ und spannend, aber auch stressig wegen der Fristen.
Wie kommen Sie zur Arbeit?
Mit dem Regionalzug von Landshut nach München, das dauert 40 Minuten. Oder remote per MS Teams.
Was möchten Sie in Zukunft machen?
Ich möchte mich auf Künstliche Intelligenz spezialisieren und eventuell zu einer größeren Firma wechseln.

Teil 3: Zusammen planen (~3-4 min dialogue)
--------------------------------------------
Typical task: "Planen Sie mit Ihrem Partner ein Team-Meeting über ein Data-Projekt."
– Discuss: Wann? Wo? Themen? Dauer? Kosten? React/agree/decide.
Ja, lass uns ein Team-Meeting für unser Data-Projekt planen.
Wann passt es, nächsten Montag um 10 Uhr?
Wo treffen wir uns – online via Teams, weil ich aus Landshut komme, oder im Büro in München?
Wir besprechen den Projektfortschritt, neue Python-Modelle und Probleme mit Daten.
Das dauert eine Stunde, und jeder bringt Kaffee mit – kostet nichts.
Gute Idee! Oder freitags um 14 Uhr?
Freitag ist besser, dann machen wir Montag 10 Uhr online.
Einverstanden? Was denkst du zu den Themen?
Perfekt, lass uns die Agenda per E-Mail schicken.

Writing: Email (~150 words)
---------------------------
Typical task: Deine Freundin fragt nach deinem neuen Job in München. Schreibe eine E-Mail (4 Punkte: Beschreibe Job; Vorteile/Nachteile; Täglicher Ablauf; Frage nach ihrem Job).​
Liebe Anna,
wie geht es dir? Vielen Dank für deine Nachricht. Ich freue mich, dir von meinem neuen Job zu erzählen.
Ich arbeite als Data Science Berater in einer Firma in München. Wir analysieren große Datenmengen und bauen Modelle mit Python.
Einerseits gefällt mir der Job sehr, weil ich kreativ arbeite und gute Gehälter habe. Andererseits sind die Deadlines manchmal stressig, und ich sitze viel am Computer.
Ein typischer Tag beginnt um 9 Uhr: Ich checke E-Mails, analysiere Daten vormittags, esse mit Kollegen zu Mittag und habe Nachmittags-Meetings. Um 17 Uhr gehe ich nach Hause.
Erzähl mir von deinem Beruf! Was machst du so und gefällt es dir?
Ich freue mich auf deine Antwort!
Liebe Grüße,
Lovely

Word count: 152. Practice timing: 30 min write, speak aloud 2x.

Gesundheit
-----------
Speaking: Teil 1 (Kennenlernen)
-------------------------------
Hallo, mein Name ist Lovely.
Ich wohne in Landshut in der Nähe von München.
Ich interessiere mich sehr für Gesundheitsthemen.
Manchmal fühle ich mich müde oder habe Kopfschmerzen, besonders wenn ich gestresst bin.
Ich versuche, gesund zu leben, indem ich viel Wasser trinke und mich bewege.
Was machen Sie, um gesund zu bleiben?
Wie kümmern Sie sich um Ihre Gesundheit?

Teil 2 : Follow-up questions with answers
---------------------------------------------
Was tun Sie, wenn Sie krank sind?
Ich ruhe mich aus, trinke viel Tee und gehe manchmal zum Arzt.
Welche Tipps haben Sie, um gesund zu bleiben?
Ich empfehle ausgewogene Ernährung, regelmäßigen Sport und genug Schlaf.
Haben Sie einen Lieblingssport?
Ja, ich mache gerne Yoga und gehe spazieren.
Wie oft trainieren Sie?
Etwa dreimal pro Woche für eine Stunde.
Wie gehen Sie mit Stress um?
Ich meditiere und mache Pausen während der Arbeit.
Was machen Sie, wenn Sie sich schlecht fühlen?
Ich bleibe zu Hause und ruhe mich aus, damit ich schnell gesund werde.

Speaking: Teil 3 (Gemeinsam planen)
------------------------------------
Lass uns überlegen, wie wir gesund im Büro bleiben können.
Sollten wir eine Sportgruppe gründen oder tägliche Pausen machen?
Wann ist die beste Zeit dafür – morgens oder nachmittags?
Wir könnten auch gesunde Snacks anbieten.
Das dauert vielleicht zehn Minuten.
Eine gute Idee!
Oder lieber zweimal pro Woche?
Einverstanden, dann machen wir zuerst ein Treffen am Montag.

Writing: E-Mail (~150 Wörter)
-------------------------------
wie geht es dir? Ich möchte dir von meiner Gesundheit erzählen.
Manchmal fühle ich mich müde oder habe Kopfschmerzen, besonders wenn ich Stress habe.
Ich versuche, gesund zu bleiben, indem ich viel Wasser trinke, gesund esse und regelmäßig Yoga mache.
Außerdem gehe ich oft spazieren und versuche, genug zu schlafen.
Wenn ich krank bin, ruhe ich mich aus und gehe zum Arzt, wenn es schlimmer wird.
Hast du Tipps, wie ich mich noch besser um meine Gesundheit kümmern kann?
Ich freue mich auf deine Antwort!
Liebe Grüße,
Lovely

Word count: approx. 150. Practice speaking Teil 1 with follow-ups calmly within 1 min, Teil 2 about 2 min, and write the email with  own ideas in 30 min.
